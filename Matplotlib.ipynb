## Matplotlib
Matplotlibは、2次元や3次元のデータを使って、折れ線グラフや棒グラフ、円グラフなどさまざまなグラフを描画し、データを可視化するためのライブラリです。
Matplotlibでグラフをビューする際のコードの書き方は、MATLABスタイルとオブジェクト指向スタイルの2種類があります。

オブジェクト指向スタイル
オブジェクト指向スタイルでは、描画オブジェクト（Figureオブジェクト）と呼ばれるグラフを描画する大枠を用意し、その中にサブプロット（Axesオブジェクト）と呼ばれるグラフを描画する領域を追加します。
そして、追加したサブプロットに対して、どのようなデータを使用するかやどのグラフで描画するかなどを指定します。

```python
# 2つのサブプロットを作成して描く
x1 = [1, 2, 3]
y1 = [2, 4, 6]
x2 = [1, 2, 3]
y2 = [-2, -4, -6]

# nrowsとncolsというパラメータでも指定
fig, axes = plt.subplots(2)
axes[0].plot(x1, y1)
axes[1].plot(x2, y2)
plt.show()
```

```python
# タイトル：set_title,suptitle
# 軸タイトル：set_xlabel,set_ylabel
# 凡例：labelパラメータ,legend
x1 = [1, 2, 3]
y1 = [2, 4, 6]
x2 = [1, 2, 3]
y2 = [-2, -4, -6]

# 1×2の領域を確保
fig, axes = plt.subplots(1, 2)
axes[0].plot(x1, y1, label='sample legend')
axes[1].plot(x2, y2)
axes[0].set_title('Subplot1 title')
axes[1].set_title('Subplot2 title')
axes[0].set_xlabel('X lable')
axes[0].set_ylabel('Y lable')
fig.suptitle('Graph title')
axes[0].legend(loc='lower right')
plt.show()
```

```python
# 棒グラフ（積み上げ）

x = [1, 2, 3]
y1 = [8, 2, 4]
y2 = [3, 1, 6]

fig, ax = plt.subplots()
labels = ['label 1', 'label 2', 'label 3']
# 横向き：ax.barh(x, y, tick_label=labels)
# tick_labelパラメータで各棒グラフにラベルを記載
ax.bar(x, y, tick_label=labels, label='y2')
ax.bar(x, y2, label='y2')
plt.legend()
plt.show()
```

```python
# 棒グラフ（隣合わせ）
fig, ax = plt.subplots()

x = [1, 2, 3]
y1 = [4, 12, 5]
y2 = [8, 2, 4]
labels = ['Apple', 'Banana', 'Orange']
width = 0.4

ax.bar(x, y1, width=width, label='y1')

x2 = [num + width for num in x]
ax.bar(x2, y2, width=width, label='y2')
ax.legend()

label_axis = [num + width/2 for num in x]
plt.xticks(label_axis, labels)

plt.show()
```

```python
# 積み上げ棒グラフ(3つ)
import numpy as np

fig, ax = plt.subplots()

x = [1, 2, 3]
y1 = [4, 12, 5]
y2 = [8, 2, 4]
y3 = [2, 6, 8]
y1_y2_total = np.add(y1, y2)
y_total = np.add(y1_y2_total, y3)
labels = ['Apple', 'Banana', 'Orange']

ax.bar(x, y_total, tick_label=labels, label='y3')
ax.bar(x, y1_y2_total, tick_label=labels, label='y2')
ax.bar(x, y1, label='y1')
ax.legend()
plt.show()
```

```python
# ヒストグラム

# data set
np.random.seed(123)
mu = 40
sigma = 5
x = np.random.normal(mu, sigma, 1000)
age_list = [round(age) for age in x]
# histgram
fig, ax = plt.subplots()
ax.set_xlabel('age')
ax.set_ylabel('quantity')
n, bins, patches = ax.hist(x, bins=30, orientation='horizontal')
plt.show()
```

```python
# 度数分布表
# n:各ビンの要素数, bins:ビンの境界値, patches:ビンを描画するためのオブジェクト情報
for i, num in enumerate(n):
    print('{:.2f} - {:.2f}: {}'.format(bins[i], bins[i + 1], num))
```

```python
# 箱ひげ図
np.random.seed(123)
x1 = np.random.normal(0, 10, 1000)
x2 = np.random.normal(0, 15, 1000)
x3 = np.random.normal(0, 20, 1000)

fig, ax = plt.subplots()
labels = ['x1', 'x2', 'x3']
ax.boxplot((x1, x2, x3), labels=labels, vert=False)
plt.show()
```

```python
# 円グラフ
df = pd.DataFrame([['テレビ、ラジオや新聞等の広告', 5],
                   ['インターネット検索結果', 15],
                   ['インターネット上の広告', 25],
                   ['SNS広告', 50],
                   ['知人からの紹介', 1],
                   ['その他', 4]], columns = ["媒体", "集計結果（%）"])

labels = df.loc[:, "媒体"]
rate = df.loc[:, "集計結果（%）"]

fig, ax = plt.subplots()
explode=[0.2, 0, 0, 0, 0, 0]
ax.pie(rate, labels=labels, startangle=90, counterclock=False, explode=explode)
# 描画するグラフのアスペクト比（グラフの縦と横の長さの比）を設定
ax.axis('equal')
plt.show()
```
