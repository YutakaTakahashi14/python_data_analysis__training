## scikit-learn
scikit-learnは、Pythonで機械学習を行うためのライブラリです。
scikit-learnを使用することで、データマイニング（統計学や人工知能などの分析手法を駆使して、データからパターンを見出すための分析手法）やデータ解析といった処理を簡単に実行できます。

代表的な前処理
*   欠損値への対応
*   カテゴリ変数の数値化
*   特徴量の正規化



```python
# 準備
import numpy as np
import pandas as pd
df = pd.DataFrame(
  {
    'A': [1, 2, 3, 4, np.nan],
    'B': [10, np.nan, 30, 40, 50],
    'C': [np.nan, 200, 300, 400, 500]
  }
)
df
```

```python
# 欠損値の確認
df.isnull()
```

```python
# 欠損値削除
df.dropna()
```

```python
# 欠損値への対応
from sklearn.impute import SimpleImputer
imp = SimpleImputer(strategy='mean')
imp.fit(df)
imp.transform(df)
```

```python
# 準備
df = pd.DataFrame([[23, 'male'],
                   [43, 'female'],
                   [31, 'male'],
                   [35, 'female'],
                   [40, 'male']],
                   columns=['age', 'sex'])
df
```

```python
# ラベルエンコーディング
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(df['sex'])
le.transform(df['sex'])
```

```python
# 準備
df = pd.DataFrame([[23, 'male'],
                   [43, 'female'],
                   [31, 'male'],
                   [35, 'female'],
                   [40, 'male']],
                   columns=['age', 'sex'])
df
```

```python
# One-hotエンコーディング
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ohe = ColumnTransformer([("OneHotEncoder", OneHotEncoder(), [1])], remainder = 'passthrough')
df_ohe = df.copy()
ohe.fit_transform(df_ohe)
```

```python
# 準備
import pandas as pd
df = pd.DataFrame({
  'data1': [1, 2, 3, 4, 5],
  'data2': [1000, 3000, 3500, 5000, 7000]
})
df
```

```python
# 分散正規化：特徴量の平均値が0、標準偏差が1になるよう特徴量を変換する
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit_transform(df)
```

```python
# 最小最大正規化：特徴量の最大値が1、最小値が0になるように特徴量を変換する
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit_transform(df)
```
